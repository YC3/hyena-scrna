# Data Preprocessing

```python
from scgpt import preprocess
```

```python

for file in ['../data/SRX9777399_ct.h5ad', '../data/SRX9856815_ct.h5ad']:
    print(file)
    # read in data
    adata = sc.read_h5ad(file)

    for layer in ['X', 'spliced' ,'unspliced']:
        preprossecor = preprocess.Preprocessor(
            use_key = layer,
            filter_gene_by_counts = False,
            filter_cell_by_counts = False,
            normalize_total = 1e4,
            result_normed_key = f"{layer}_normed",
            log1p = True,
            result_log1p_key = f"{layer}_log1p",
            subset_hvg = False,
            hvg_use_key = f"{layer}_normed", 
            binning = 15,
            result_binned_key = f"{layer}_binned"
        )
        
        # inplace
        preprossecor(adata)


    adata.write_h5ad(f"../data/{file.split('/')[-1].split('_ct.h5ad')[0]}_binned.h5ad")
```
